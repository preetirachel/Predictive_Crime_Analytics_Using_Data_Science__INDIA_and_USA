```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd



data = pd.read_csv('01_District_wise_crimes_committed_IPC_2001_2012a.csv')

```

```python
data = pd.read_csv('01_District_wise_crimes_committed_IPC_2001_2012a.csv')

X = data[["ATTEMPT TO MURDER","MURDER"]]
#Visualise data points
plt.scatter(X["MURDER"],X["ATTEMPT TO MURDER"],c='black')
plt.xlabel('MURDER')
plt.ylabel('ATTEMPT TO MURDER')
plt.show()
```

```python
#number of clusters
K=3

# Select random observation as centroids
Centroids = (X.sample(n=K))
plt.scatter(X["MURDER"],X["ATTEMPT TO MURDER"],c='black')
plt.scatter(Centroids["MURDER"],Centroids["ATTEMPT TO MURDER"],c='red')
plt.xlabel('MURDER')
plt.ylabel('ATTEMPT TO MURDER')
plt.show()
```

```python
# Step 3 - Assign all the points to the closest cluster centroid
# Step 4 - Recompute centroids of newly formed clusters
# Step 5 - Repeat step 3 and 4

diff = 1
j=0

while(diff!=0):
    XD=X
    i=1
    for index1,row_c in Centroids.iterrows():
        ED=[]
        for index2,row_d in XD.iterrows():
            d1=(row_c["MURDER"]-row_d["MURDER"])**2
            d2=(row_c["ATTEMPT TO MURDER"]-row_d["ATTEMPT TO MURDER"])**2
            d=np.sqrt(d1+d2)
            ED.append(d)
        X[i]=ED
        i=i+1

    C=[]
    for index,row in X.iterrows():
        min_dist=row[1]
        pos=1
        for i in range(K):
            if row[i+1] < min_dist:
                min_dist = row[i+1]
                pos=i+1
        C.append(pos)
    X["Cluster"]=C
    Centroids_new = X.groupby(["Cluster"]).mean()[["ATTEMPT TO MURDER","MURDER"]]
    if j == 0:
        diff=1
        j=j+1
    else:
        diff = (Centroids_new['ATTEMPT TO MURDER'] - Centroids['ATTEMPT TO MURDER']).sum() + (Centroids_new['MURDER'] - Centroids['MURDER']).sum()
        print(diff.sum())
    Centroids = X.groupby(["Cluster"]).mean()[["ATTEMPT TO MURDER","MURDER"]]
```

```python
color=['blue','green','cyan']
for k in range(K):
    data=X[X["Cluster"]==k+1]
    plt.scatter(data["MURDER"],data["ATTEMPT TO MURDER"],c=color[k])
plt.scatter(Centroids["MURDER"],Centroids["ATTEMPT TO MURDER"],c='red')
plt.xlabel('MURDER')
plt.ylabel('ATTEMPT TO MURDER')
plt.show()
```

```python
df = pd.read_csv('01_District_wise_crimes_committed_IPC_2013.csv')
```

```python
X = df[['RAPE','CRUELTY BY HUSBAND OR HIS RELATIVES','INSULT TO MODESTY OF WOMEN','ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY']]

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform( X )
cluster_range = range( 1, 20 )
cluster_errors = []

for num_clusters in cluster_range:
  clusters = KMeans( num_clusters )
  clusters.fit(X_scaled)
  cluster_errors.append( clusters.inertia_ )

clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )

clusters_df[0:10]
```

```python
plt.figure(figsize=(16,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" );
```

```python
X = df[['RAPE','CRUELTY BY HUSBAND OR HIS RELATIVES','INSULT TO MODESTY OF WOMEN','ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY']]

clusters = KMeans(4)  # 4 clusters!
clusters.fit( X )
clusters.cluster_centers_
clusters.labels_

df['Crime_clusters'] = clusters.labels_
df.head()
df.sort_values(by=['Crime_clusters'],ascending = True)
X.head()
```

```python
sns.lmplot('MURDER' ,'RAPE', data=df,
        hue = "Crime_clusters",
        fit_reg=False, size = 6 );
```

```python
sns.lmplot( 'INSULT TO MODESTY OF WOMEN','RAPE',  data=df,
        hue = 'Crime_clusters',
        fit_reg=False, size = 6 );
```
