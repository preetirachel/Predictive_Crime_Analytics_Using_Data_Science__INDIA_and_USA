```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd



data = pd.read_csv('01_District_wise_crimes_committed_IPC_2001_2012a.csv')

```

```python
data.head()
```

```python
X = data[["RAPE","YEAR"]]
#Visualise data points
plt.scatter(X["YEAR"],X["RAPE"],c='black')
plt.xlabel('YEAR')
plt.ylabel('RAPE')
plt.show()
```

```python
#number of clusters
K=3

# Select random observation as centroids
Centroids = (X.sample(n=K))
plt.scatter(X["YEAR"],X["RAPE"],c='black')
plt.scatter(Centroids["YEAR"],Centroids["RAPE"],c='red')
plt.xlabel('YEAR')
plt.ylabel('RAPE')
plt.show()
```

```python
diff = 1
j=0

while(diff!=0):
    XD=X
    i=1
    for index1,row_c in Centroids.iterrows():
        ED=[]
        for index2,row_d in XD.iterrows():
            d1=(row_c["YEAR"]-row_d["RAPE"])**2
            d2=(row_c["RAPE"]-row_d["RAPE"])**2
            d=np.sqrt(d1+d2)
            ED.append(d)
        X[i]=ED
        i=i+1

    C=[]
    for index,row in X.iterrows():
        min_dist=row[1]
        pos=1
        for i in range(K):
            if row[i+1] < min_dist:
                min_dist = row[i+1]
                pos=i+1
        C.append(pos)
    X["Cluster"]=C
    Centroids_new = X.groupby(["Cluster"]).mean()[["RAPE","YEAR"]]
    if j == 0:
        diff=1
        j=j+1
    else:
        diff = (Centroids_new['RAPE'] - Centroids['RAPE']).sum() + (Centroids_new['YEAR'] - Centroids['YEAR']).sum()
        print(diff.sum())
    Centroids = X.groupby(["Cluster"]).mean()[["RAPE","YEAR"]]
```

```python
color=['blue','green','cyan']
for k in range(K):
    data=X[X["Cluster"]==k+1]
    plt.scatter(data["YEAR"],data["RAPE"],c=color[k])
plt.scatter(Centroids["YEAR"],Centroids["RAPE"],c='red')
plt.xlabel('YEAR')
plt.ylabel('RAPE')
plt.show()
```

```python
data = pd.read_csv('01_District_wise_crimes_committed_IPC_2001_2012a.csv')

X = data[["ATTEMPT TO MURDER","MURDER"]]
#Visualise data points
plt.scatter(X["MURDER"],X["ATTEMPT TO MURDER"],c='black')
plt.xlabel('MURDER')
plt.ylabel('ATTEMPT TO MURDER')
plt.show()
```

```python
#number of clusters
K=8

# Select random observation as centroids
Centroids = (X.sample(n=K))
plt.scatter(X["MURDER"],X["ATTEMPT TO MURDER"],c='black')
plt.scatter(Centroids["MURDER"],Centroids["ATTEMPT TO MURDER"],c='red')
plt.xlabel('MURDER')
plt.ylabel('ATTEMPT TO MURDER')
plt.show()
```

```python
# Step 3 - Assign all the points to the closest cluster centroid
# Step 4 - Recompute centroids of newly formed clusters
# Step 5 - Repeat step 3 and 4

diff = 1
j=0

while(diff!=0):
    XD=X
    i=1
    for index1,row_c in Centroids.iterrows():
        ED=[]
        for index2,row_d in XD.iterrows():
            d1=(row_c["MURDER"]-row_d["MURDER"])**2
            d2=(row_c["ATTEMPT TO MURDER"]-row_d["ATTEMPT TO MURDER"])**2
            d=np.sqrt(d1+d2)
            ED.append(d)
        X[i]=ED
        i=i+1

    C=[]
    for index,row in X.iterrows():
        min_dist=row[1]
        pos=1
        for i in range(K):
            if row[i+1] < min_dist:
                min_dist = row[i+1]
                pos=i+1
        C.append(pos)
    X["Cluster"]=C
    Centroids_new = X.groupby(["Cluster"]).mean()[["ATTEMPT TO MURDER","MURDER"]]
    if j == 0:
        diff=1
        j=j+1
    else:
        diff = (Centroids_new['ATTEMPT TO MURDER'] - Centroids['ATTEMPT TO MURDER']).sum() + (Centroids_new['MURDER'] - Centroids['MURDER']).sum()
        print(diff.sum())
    Centroids = X.groupby(["Cluster"]).mean()[["ATTEMPT TO MURDER","MURDER"]]
```

```python
color=['blue','green','cyan']
for k in range(K):
    data=X[X["Cluster"]==k+1]
    plt.scatter(data["MURDER"],data["ATTEMPT TO MURDER"],c=color[k])
plt.scatter(Centroids["MURDER"],Centroids["ATTEMPT TO MURDER"],c='red')
plt.xlabel('MURDER')
plt.ylabel('ATTEMPT TO MURDER')
plt.show()
```
